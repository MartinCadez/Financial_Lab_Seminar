```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
plt.rcParams["font.family"] = "times new roman"
plt.rcParams["font.sans-serif"] = ["Times New Roman"]

G = nx.Graph()
G.add_edges_from([("u", "v_1"), ("u", "v_2")])

pos = {
    "u": (0, 0),
    "v_1": (0.9, 0),
    "v_2": (1.8, 0),
}

nx.draw(
    G,
    pos,
    with_labels=True,
    labels={"u": "u", "v_1": r"$v_1$", "v_2": r"$v_2$"},
    node_color=["#698ad1", "#b5dcff", "#b5dcff"],
    node_size=4000,
    font_color="black",
    edge_color="#292a40",
    width=2,
    linewidths=1.1,
    edgecolors="#292a40",
    font_size=30,
)

plt.xlim(-0.3, 2.1)
plt.ylim(-0.3, 0.3)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
```

```python
plt.rcParams["font.family"] = "times new roman"
G = nx.Graph()
edges = [
    ("v_1", "v_2"),
    ("v_1", "v_3"),
    ("v_1", "v_5"),
    ("v_2", "v_3"),
    ("v_2", "v_4"),
    ("v_3", "v_4"),
    ("v_4", "v_5"),
    ("v_3", "v_5"),
]

G.add_edges_from(edges)

pos = {
    "v_1": (0, 1),
    "v_2": (1, 1),
    "v_3": (0.5, 0.5),
    "v_4": (1, 0),
    "v_5": (0, 0),
}

node_colors = {
    "v_1": "#b5dcff",
    "v_2": "#698ad1",
    "v_3": "#698ad1",
    "v_4": "#b5dcff",
    "v_5": "#698ad1",
}

node_color_list = [node_colors[node] for node in G.nodes()]

nx.draw(
    G,
    pos,
    with_labels=True,
    labels={
        "v_1": r"$v_1$",
        "v_2": r"$v_2$",
        "v_3": r"$v_3$",
        "v_4": r"$v_4$",
        "v_5": r"$v_5$",
    },
    node_color=node_color_list,
    node_size=3200,
    font_color="black",
    edge_color="#292a40",
    width=2,
    linewidths=2,
    edgecolors="#292a40",
    font_size=30,
)

plt.gca().set_facecolor("#000000")
plt.xlim(-0.3, 1.3)
plt.ylim(-0.2, 1.2)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
```
